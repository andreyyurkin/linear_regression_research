```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../datasets/coffee_shop_revenue.csv")
```

```python
df
```

```python
plt.title('target dist')
plt.grid(which='major')
sns.histplot(df.Daily_Revenue, kde=True)
plt.show()
```

```python
plt.title("corr matrix")
sns.heatmap(df.corr())
plt.show()
```

```python
target = "Daily_Revenue"
features = list(df.columns)
features.remove(target)
```

```python
features
```

```python
len(features)
```

```python
for feature in features:
    plt.title(f"feature dist - {feature}")
    sns.histplot(df[feature])
    plt.show()
```

```python
features[0]
```

```python
sns.lineplot(data=df, y=target, x=features[0])
```
